```python
# Data audit for application_record.csv
# Check for missing values in each column
missing_values = application_record_pd.isnull().sum()
print("Missing values in each column in application_record.csv:\n", missing_values)
print("=================================================")

# Check for duplicates based on the 'ID' column
duplicate_ids = application_record_pd.duplicated(subset='ID').sum()
print("Number of duplicate IDs in application_record.csv:", duplicate_ids)
print("=================================================")

# Detect outliers using the IQR
def detect_and_count_outliers(df):
    outlier_counts = {}
    numeric_columns = df.select_dtypes(include=['int64','float64']).columns
    for column in numeric_columns:
        unique_values = df[column].unique()
        if len(unique_values) <= 3:
            continue
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers = df[column][(df[column] < lower_bound) | (df[column] > upper_bound)]
        outlier_counts[column] = outliers.count()
    return outlier_counts

print("Outliers in each column in application record:")
outliers_result = detect_and_count_outliers(application_record_pd)
for column, count in outliers_result.items():
    print(f"{column}: {count}")
```

```python
# Data audit for credit_record.csv
# Check for missing values in each column
missing_values = credit_record_pd.isnull().sum()
print("Missing values in each column in credit_record.csv:\n", missing_values)
print("=================================================")

# Check for complete duplicate rows
duplicate_rows = credit_record_pd.duplicated().sum()
print("Number of duplicate rows in credit_record.csv:", duplicate_rows)
print("=================================================")

# Detect outliers using the IQR
outliers_result = detect_and_count_outliers(credit_record_pd)
print("Outliers in each column in credit record:")
for column, count in outliers_result.items():
    print(f"{column}: {count}")
```

```python
# Drop unnecessary columns
application_record_pd = application_record_pd.drop(columns=['FLAG_PHONE', 'FLAG_EMAIL','FLAG_MOBIL'])
```

```python
# Drop duplicate rows based on 'ID'
application_record_pd = application_record_pd.drop_duplicates(subset='ID', keep='first')
```

```python
import pandas as pd
# Handle missing OCCUPATION_TYPE
def update_occupation_type(row):
    if pd.isna(row['OCCUPATION_TYPE']):
        if row['DAYS_EMPLOYED'] > 0:
            return 'not working'
        else:
            return 'unknown'
    else:
        return row['OCCUPATION_TYPE']

application_record_pd['OCCUPATION_TYPE'] = application_record_pd.apply(update_occupation_type, axis=1)
print(application_record_pd[['ID','OCCUPATION_TYPE', 'DAYS_EMPLOYED']].head(1532))
```

```python
import numpy as np
# Log transformation of AMT_INCOME_TOTAL
temp = np.log1p(application_record_pd['AMT_INCOME_TOTAL'])

plt.figure(figsize=(10, 6))
plt.hist(temp, bins=50, color='green', alpha=0.7)
plt.xlabel('AMT_INCOME_TOTAL_LogN')
plt.ylabel('COUNT')
plt.grid(True)
plt.show()
```

```python
# Construct classification from credit_record
constructed_credit_record = pd.DataFrame()
constructed_credit_record['ID'] = credit_record_pd['ID'].unique()

def classify_client(id):
    client_status = credit_record_pd[credit_record_pd['ID'] == id]['STATUS'].tolist()
    has_bad_status = any(status in ['1', '2', '3', '4', '5'] for status in client_status)
    if has_bad_status:
        return 0
    all_history_insufficient = all(status == 'X' for status in client_status)
    if all_history_insufficient:
        return -1
    return 1

constructed_credit_record['classification'] = constructed_credit_record['ID'].map(classify_client)
classification_counts = constructed_credit_record['classification'].value_counts()
classification_counts.sort_index().plot(kind='bar', color='skyblue')
plt.title('Distribution of Classification')
plt.xlabel('Classification')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1, 2], labels=['-1: Insufficient History', '0: Bad Payer', '1: Low-risk Clients'], rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Merge Data
merged_data = pd.merge(application_record_pd, constructed_credit_record, on='ID', how='inner')
```

```python
# Transform all categorical columns to numerical
categorical_columns = ['CODE_GENDER','FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 
                       'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 
                       'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE', 
                       'OCCUPATION_TYPE']

categorical_data = merged_data[categorical_columns]

def factorize_columns(data):
    numerical_data = {col: pd.factorize(data[col])[0] for col in data}
    return pd.DataFrame(numerical_data)

numerical_categorical_data = factorize_columns(categorical_data)

for column in categorical_columns:
    merged_data[column] = numerical_categorical_data[column]
```

```python
# Reduce Data
merged_data = merged_data[merged_data['classification'] != -1]
merged_data = merged_data.drop(columns=['ID'])
```

```python
# Log transformation of AMT_INCOME_TOTAL in merged data
temp = np.log1p(merged_data['AMT_INCOME_TOTAL'])

plt.figure(figsize=(10, 6))
plt.hist(temp, bins=50, color='red', alpha=0.7)
plt.xlabel('AMT_INCOME_TOTAL_LogN')
plt.ylabel('COUNT')
plt.grid(True)
plt.show()
```

```python
merged_data.info()
# Convert merged_data back to Spark DataFrame
merged_data_spark = spark.createDataFrame(merged_data)

from pyspark.ml.feature import VectorAssembler
# Assemble features into a vector
assembler = VectorAssembler(inputCols=merged_data_spark.columns[:-1], outputCol='features')
final_data = assembler.transform(merged_data_spark).select('features', 'classification')
```

```python
# Partition
train_data, test_data = final_data.randomSplit([0.7, 0.3], seed=0)
```

```python
# Train the RandomForestClassifier
from pyspark.ml.classification import RandomForestClassifier
rf = RandomForestClassifier(labelCol='classification', featuresCol='features', numTrees=100, seed=0)
model = rf.fit(train_data)


# Make predictions
predictions = model.transform(test_data)


# Evaluate the model
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
evaluator = MulticlassClassificationEvaluator(labelCol='classification', predictionCol='prediction', metricName='accuracy')
accuracy = evaluator.evaluate(predictions)
print(f'Testing Accuracy: {accuracy * 100:.2f}%')
```

```python
# Evaluate the model using various metrics
accuracy = evaluator.evaluate(predictions, {evaluator.metricName: "accuracy"})
weighted_precision = evaluator.evaluate(predictions, {evaluator.metricName: "weightedPrecision"})
weighted_recall = evaluator.evaluate(predictions, {evaluator.metricName: "weightedRecall"})
f1_score = evaluator.evaluate(predictions, {evaluator.metricName: "f1"})

print(f"Testing Accuracy: {accuracy * 100:.2f}%")
print(f"Weighted Precision: {weighted_precision * 100:.2f}%")
print(f"Weighted Recall: {weighted_recall * 100:.2f}%")
print(f"F1 Score: {f1_score * 100:.2f}%")
```

```python
from pyspark.ml.evaluation import BinaryClassificationMetrics# 
predictionAndLabels = predictions.select("prediction", "classification")

# RDD
predictionAndLabelsRDD = predictionAndLabels.rdd.map(lambda row: (row.prediction, row.classification))

# MulticlassMetrics
metrics = MulticlassMetrics(predictionAndLabelsRDD)

# 
confusionMatrix = metrics.confusionMatrix()
```
